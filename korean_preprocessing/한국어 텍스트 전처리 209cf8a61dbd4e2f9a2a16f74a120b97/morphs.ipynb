# 형태소 분석기 비교 - okt, mecab, komoran, kiwi

```python
# konlpy내 okt(시중에 코드가 제일 많음), mecab(제일 좋음)
from konlpy.tag import Okt, Mecab, Komoran

Okt = Okt()
mecab = Mecab()
komoran = Komoran()

# 카카오에서 만든 kiwi형태소 분석기
from kiwipiepy import Kiwi
from kiwipiepy.utils import Stopwords

# kiwi 사용하기
kiwi = Kiwi()
kiwi.prepare() # 전처리 위한 준비
stopwords = Stopwords() # 불용어 사전 불러오기

# 예시 문장
sentence1 = "낮에는 밥을 먹고 밤엔 밤을 먹었다." 
sentence2 = "나는어제도대우관별관이었는데오늘도대우관별관이다."
sentence3 = "짝남이랑썸붕나서어제너무힘들어따ㅠㅠ"
```

### 1. 형태소 분석만 해보기

```python
# (형태소, 품사) 형태로 출력
print('---------okt 형태소 분석---------')
print(Okt.pos(sentence1))

print('---------mecab 형태소 분석---------')
print(mecab.pos(sentence1))

print('---------komoran 형태소 분석---------')
print(komoran.pos(sentence1))
```

```python
# kiwi 형태소 분석
print('---------kiwi 형태소 분석---------')
kiwi.tokenize(sentence1)
```

```python
# (형태소, 품사) 형태로 출력
print('---------okt 형태소 분석---------')
print(Okt.pos(sentence2))

print('---------mecab 형태소 분석---------')
print(mecab.pos(sentence2))

print('---------komoran 형태소 분석---------')
print(komoran.pos(sentence2))
```

```python
# kiwi 형태소 분석
print('---------kiwi 형태소 분석---------')
kiwi.tokenize(sentence2)
```

### 2. 원형복원까지

```python
# okt
tokens_okt= Okt.morphs(sentence1, norm = True, stem = True) #norm: 표준화
print('---------okt 원형복원---------')
print(tokens_okt)

# mecab
tokens_mecab = mecab.pos(sentence1)
tok_mecab = [word + ('다' if tag.startswith('V') else '') for word, tag in tokens_mecab]
print('---------mecab 원형복원---------')
print(tok_mecab)

# komoran
tokens_komoran = komoran.pos(sentence1)
tok_komoran = [word + ('다' if tag.startswith('V') else '') for word, tag in tokens_komoran]
print('---------mecab 원형복원---------')
print(tok_komoran)

# kiwi
tokens = kiwi.tokenize(sentence1)# stopwords = stopwords
morph_list = [word + ('다' if tag.startswith('V') else '') for word, tag, _, _ in tokens]
print('---------kiwi 원형복원---------')
print(morph_list)
```
